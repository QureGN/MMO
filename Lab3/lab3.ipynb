```python
K2 = 1.5
IQR = df['avg_glucose_level'].quantile(0.75) - df['avg_glucose_level'].quantile(0.25)
lower_boundary = df['avg_glucose_level'].quantile(0.25) - (K2 * IQR)
upper_boundary = df['avg_glucose_level'].quantile(0.75) + (K2 * IQR)
```

```python
lower_boundary, upper_boundary 
```

```python
def diagnostic_plots(df, variable, title):
    fig, ax = plt.subplots(figsize=(10,7))
    # гистограмма
    plt.subplot(2, 2, 1)
    df[variable].hist(bins=30)
    ## Q-Q plot
    plt.subplot(2, 2, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    # ящик с усами
    plt.subplot(2, 2, 3)
    sns.violinplot(x=df[variable])    
    # ящик с усами
    plt.subplot(2, 2, 4)
    sns.boxplot(x=df[variable])  
    fig.suptitle(title)
    plt.show()
```

```python
diagnostic_plots(df[['avg_glucose_level']], 'avg_glucose_level' , "r")
```

# Один метод из группы методов обертывания (wrapper methods);

Алгоритмы полного перебора

```python
pip install mlxtend
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
X = df[['precipitation', 'temp_max', 'temp_min', 'wind' ]]
y = df['weather']
```

```python
efs1 = EFS(knn, 
           min_features=2,
           max_features=4,
           scoring='accuracy',
           print_progress=True,
           cv=4)

efs1 = efs1.fit(X, y)

print('Best accuracy score: %.2f' % efs1.best_score_)
print('Best subset (indices):', efs1.best_idx_)
print('Best subset (corresponding names):', efs1.best_feature_names_)
```
